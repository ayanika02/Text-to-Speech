```python
with open('finaldata.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['text', 'audio'])  # Header
    
    for i, sentence in enumerate(cleaned_sentences):
        if len(sentence)<1:
            continue
        else:
            audio_file = f'train_{i}.mp3'  
            writer.writerow([sentence, audio_file])

print("Dataset CSV created successfully.")
```
